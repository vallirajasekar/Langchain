```python
import bs4
```

```python
loader = WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/India", "https://gwcdata.ai/"),bs_kwargs=dict(parse_only=bs4.SoupStrainer),
                       class_=("post-title","post-content","post-header"))
```

```python
loader = WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/India", "https://gwcdata.ai/"),
                      bs_kwargs=dict(parse_only=bs4.SoupStrainer),
                      class=("post-title", "post-content", "post-header"))
```

```python
loader = WebBaseLoader(
    web_paths=("https://en.wikipedia.org/wiki/India", "https://gwcdata.ai/"),
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(
    class_=("post-title", "post-content", "post-header")
))
)
```

```python
loader.load()
```
